```python
import pandas as pd
import seaborn as sns        # also python's library for visualisations
color = sns.color_palette()
sns.set_style('darkgrid')

import matplotlib            # python's library for visualisation
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('databrasilescola_skills.csv')
```

```python
sns.distplot(df_train.finalGrade) 
plt.ylabel('Frequency')
plt.savefig('finalgradeDatabrasil.png', dpi=1200)
```

```python
df_uol = pd.read_csv('datav2_skills.csv')
```

```python
sns.distplot(df_uol.finalGrade) 
plt.ylabel('Frequency')
plt.savefig('finalgradeDataUol.png', dpi=1200)   
```

```python
print(df_train.finalGrade.mean())
print(df_train.finalGrade.std())
```
